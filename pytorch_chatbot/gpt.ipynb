```python
import torch
import torch.nn as nn
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
#import training text
with open("input.txt", 'r', encoding="utf-8") as f:
    input_text = f.read()
```

```python
#"tokenizing" text into chunks of 1 char and then simple integer encoding
chars = sorted(list(set(input_text)))
vocab_size = len(chars)
#print("".join(chars))
#print(vocab_size)

string_to_int = {ch:i for i, ch in enumerate(chars)}
int_to_string = {i:ch for i, ch in enumerate(chars)}
def encode(sentence):
    return [string_to_int[ch] for ch in sentence]
def decode(sequence):
    return [int_to_string[i] for i in sequence]

tokenized_input = encode(input_text)
```

```python
input_data = torch.torch(tokenized_input, dtype=torch.long)
```
